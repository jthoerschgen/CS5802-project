```python
import pandas as pd

import random

random.seed(794)
```

```python
names = pd.read_csv("names.csv", header=1)
names.head(10)
```

```python
display(names[names.duplicated()])
assert not names[names.duplicated()].shape[0]
```

```python
departments = pd.read_csv("courses.csv")
departments.head(10)
```

```python
semesters = []
for year in [2020, 2021, 2022, 2023, 2024]:
    for term in ["SP", "FS"]:
        semesters.append(f"{term}{year}")
```

```python
data = []
# for _, name in names.iterrows():
for student_id in range(0, 80 +1, 1):
    all_student_courses = set()  # keep track of courses

    student = {
        # "Name": f"{name['First Name']} {name['Last Name']}",
        "StudentID": f"{student_id}",
        "Semesters": [],
    }
    for i, term in enumerate(semesters):
        semester = {"Semester": term, "Courses": []}

        # create example courses
        credit_hours = []
        while sum(credit_hours) < 15:
            course_number = max(
                (i // 2) + 1, ((i // 2) + 1) + random.choice([-1, 0, 1])
            )
            course_number = course_number * 1000 + random.randint(2, 999)
            course_abbreviation = random.choice(departments["Abbreviation"])
            hours = random.choices([2, 3, 4], weights=[15, 65, 20], k=1)[0]
            grade = random.choices(
                ["A", "B", "C", "D", "F"], weights=[20, 35, 30, 8, 7], k=1
            )[0]

            course = {
                "Course": f"{course_abbreviation} {course_number}",
                "Hours": hours,
                "Grade": grade,
            }

            if course["Course"] in all_student_courses:  # Avoid duplicates
                print(
                    # f"Avoiding Duplicate Course: {name['First Name']} {name['Last Name']} "
                    f"Avoiding Duplicate Course: {student_id} "
                    + f"({term}),\n{course}\n"
                )
                continue
            all_student_courses.add(course["Course"])

            credit_hours.append(hours)
            semester["Courses"].append(course)
        student["Semesters"].append(semester)
    data.append(student)
```

```python
df_data = []
for student in data:
    for semester in student["Semesters"]:
        for course in semester["Courses"]:
            df_data.append(
                {
                    "StudentID": student["StudentID"],
                    "Semester": semester["Semester"],
                    "Course": course["Course"],
                    "Hours": course["Hours"],
                    "Grade": course["Grade"],
                }
            )
```

```python
grade_data = pd.DataFrame(df_data)
grade_data.head(10)
```

```python
print(grade_data.shape[0], "total data entries")
```

```python
for semester, group in grade_data.groupby(by="Semester"):
    group.sample(frac=1, random_state=42).to_csv(
        f"example_data/example_grade_data_{semester}.csv", index=False
    )
    print(f"Saving data for {semester} as CSV file ({group.size} rows)")
```
